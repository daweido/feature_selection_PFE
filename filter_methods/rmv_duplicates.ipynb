```
def rmv_duplicates(data):
  data_T = data.T
  unique_features = data_T.drop_duplicates(keep='first').T
  
  remaining_features=[dup_col for dup_col in data.columns if dup_col  in unique_features.columns]
  print(f"\nRemaining features labels : \t{remaining_features}")
  print(f"Remaining features count : {len(remaining_features)}")

  removed_features = [dup_col for dup_col in data.columns if dup_col not in unique_features.columns]
  print(f"\nRemoved features labels {removed_features}")
  print(f"Removed features count : {len(removed_features)}")
```

```
rmv_duplicates(X_train)
```
