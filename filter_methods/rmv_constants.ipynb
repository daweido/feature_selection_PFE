```
import numpy as np
from sklearn.feature_selection import VarianceThreshold
```

```
constant_filter = VarianceThreshold(threshold=0)
#setting the variance threshold to 0 to detect constant features
```

```
constant_filter.fit(X_train)
```

```
len(X_train.columns[constant_filter.get_support()])

```

**Removed features (0 in this case)**

```
constant_columns = [column for column in X_train.columns
                    if column not in X_train.columns[constant_filter.get_support()]]

print(len(constant_columns))
```

**Filter features**

```
X_train_filter=X_train.drop(labels=constant_columns, axis=1)

#Return a dataframe
```

**Function for the main script**

```
def rmv_constant(data):

  constant_filter = VarianceThreshold(threshold=0)
  constant_filter.fit(data)


  print(f"\nRemaining features labels : \t{data.columns[constant_filter.get_support()]}")
  print(f"Remaining features count : \t{len(data.columns[constant_filter.get_support()])}")
  

  constant_columns = [column for column in data.columns
                      if column not in data.columns[constant_filter.get_support()]]

  print(f"\nRemoved features labels {constant_columns}")
  print(f"Removed features count : {len(constant_columns)}")

  




```

```
rmv_constant(X_train)
```
